```python
import pandas as pd

# Load the datasets
try:
    disease_symptoms_df = pd.read_csv('DiseaseAndSymptoms.csv')
    disease_precaution_df = pd.read_csv('Disease precaution.csv')
    print("Datasets loaded successfully.")
except FileNotFoundError:
    print("Make sure 'DiseaseAndSymptoms.csv' and 'Disease precaution.csv' are in the correct directory.")
    # Create dummy dataframes for demonstration if files are not found
    data = {'symptoms': ['cough, fever', 'headache, nausea', 'skin rash, itching'],
            'diagnosis': ['common cold', 'migraine', 'allergic reaction'],
            'how_it_happens': ['viral infection', 'neurological factors', 'immune response'],
            'ayurvedic_treatment': ['ginger tea, rest', 'herbal remedies', 'neem paste'],
            'conventional_treatment': ['pain relievers, decongestants', 'pain medication, anti-nausea drugs', 'antihistamines, topical creams']}
    df = pd.DataFrame(data)
    disease_symptoms_df = pd.DataFrame({
        'Disease': ['Fungal infection', 'Fungal infection', 'GERD'],
        'Symptom_1': ['itching', 'skin_rash', 'stomach_pain'],
        'Symptom_2': ['skin_rash', 'nodal_skin_eruptions', 'acidity'],
        'Symptom_3': ['nodal_skin_eruptions', 'dischromic _patches', 'ulcers_on_tongue'],
        'Symptom_4': ['dischromic _patches', 'none', 'vomiting']
    })
    disease_precaution_df = pd.DataFrame({
        'Disease': ['Fungal infection', 'GERD'],
        'Precaution_1': ['bath twice', 'avoid fatty spicy food'],
        'Precaution_2': ['use detol or neem in bathing water', 'avoid lying down after eating'],
        'Precaution_3': ['keep infected area dry', 'maintain healthy weight'],
        'Precaution_4': ['use clean cloths', 'exercise']
    })
    print("Using dummy data due to file not found.")


# Display the first few rows of each DataFrame
print("\nDisease and Symptoms Data:")
display(disease_symptoms_df.head())

print("\nDisease Precaution Data:")
display(disease_precaution_df.head())
```

```python
# Display information about the dataframes
print("Disease and Symptoms DataFrame Info:")
disease_symptoms_df.info()

print("\nDisease Precaution DataFrame Info:")
disease_precaution_df.info()

# Check for missing values
print("\nMissing values in Disease and Symptoms DataFrame:")
print(disease_symptoms_df.isnull().sum())

print("\nMissing values in Disease Precaution DataFrame:")
print(disease_precaution_df.isnull().sum())

# Check the distribution of diseases in the symptoms dataset
print("\nDistribution of diseases in Disease and Symptoms DataFrame:")
display(disease_symptoms_df['Disease'].value_counts())

# Check if the diseases in both dataframes match
diseases_symptoms = set(disease_symptoms_df['Disease'].unique())
diseases_precaution = set(disease_precaution_df['Disease'].unique())

diseases_match = diseases_symptoms == diseases_precaution

print(f"\nDo the diseases in both dataframes match? {diseases_match}")

if not diseases_match:
    print("\nDiseases present in Disease and Symptoms but not in Disease Precaution:")
    print(list(diseases_symptoms - diseases_precaution))
    print("\nDiseases present in Disease Precaution but not in Disease and Symptoms:")
    print(list(diseases_precaution - diseases_symptoms))
```

```python
# Get all unique symptoms from the disease_symptoms_df
all_symptoms = set()
for col in disease_symptoms_df.columns:
    if col.startswith('Symptom'):
        # Fill NaN values with an empty string to avoid errors during iteration
        all_symptoms.update(disease_symptoms_df[col].dropna().unique())

# Create a list of unique symptoms
all_symptoms = sorted(list(all_symptoms))

# Create a dictionary to map symptoms to their index
symptom_to_index = {symptom: i for i, symptom in enumerate(all_symptoms)}

# Create a new DataFrame for encoded symptoms
encoded_symptoms_df = pd.DataFrame(0, index=disease_symptoms_df.index, columns=all_symptoms)

# Populate the encoded_symptoms_df
for index, row in disease_symptoms_df.iterrows():
    for col in disease_symptoms_df.columns:
        if col.startswith('Symptom'):
            symptom = row[col]
            if pd.notna(symptom):
                encoded_symptoms_df.loc[index, symptom] = 1

print("Encoded Symptoms DataFrame:")
display(encoded_symptoms_df.head())

# Prepare the target variable (Disease)
y_diseases = disease_symptoms_df['Disease']

print("\nDisease Target Variable:")
display(y_diseases.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(encoded_symptoms_df, y_diseases, test_size=0.2, random_state=42, stratify=y_diseases)

# Initialize and train the Random Forest Classifier
# Using class_weight='balanced' to handle potential class imbalance
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Print classification report for more detailed evaluation
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Create a dictionary to map diseases to their precautions
disease_precautions = disease_precaution_df.set_index('Disease').T.to_dict('list')

# Function to get precautions for a given disease
def get_precautions(disease):
    """
    Retrieves precautions for a given disease.

    Args:
        disease (str): The name of the disease.

    Returns:
        list: A list of precautions for the disease, or a message if not found.
    """
    precautions = disease_precautions.get(disease)
    if precautions:
        # Remove any None values that might be present
        return [p for p in precautions if pd.notna(p)]
    else:
        return [f"Precautions not found for {disease} in the precaution dataset."]

# Test the function with a sample disease
sample_disease = disease_precaution_df['Disease'].sample(1).iloc[0]
print(f"Precautions for {sample_disease}:")
display(get_precautions(sample_disease))
```

```python
# Function to predict disease and get precautions
def predict_disease_and_precautions(symptoms_list, model, all_symptoms, disease_precautions):
    """
    Predicts the disease based on symptoms and retrieves precautions.

    Args:
        symptoms_list (list): A list of symptoms provided by the user.
        model: The trained machine learning model.
        all_symptoms (list): A list of all unique symptoms the model was trained on.
        disease_precautions (dict): A dictionary mapping diseases to precautions.

    Returns:
        str: A formatted string containing the predicted disease and precautions.
    """
    # Create a one-hot encoded vector for the input symptoms
    input_encoded_symptoms = pd.DataFrame(0, index=[0], columns=all_symptoms)
    for symptom in symptoms_list:
        if symptom in all_symptoms:
            input_encoded_symptoms[symptom] = 1
        else:
            print(f"Warning: Symptom '{symptom}' not recognized and will be ignored.")


    # Predict the disease
    predicted_disease = model.predict(input_encoded_symptoms)[0]

    # Get precautions for the predicted disease
    precautions = get_precautions(predicted_disease)

    # Format the response
    response = f"Based on the symptoms you provided, the predicted disease is: **{predicted_disease}**\n\n"

    response += "Here are some precautions you can take:\n"
    if precautions:
        for i, precaution in enumerate(precautions):
            response += f"- {precaution}\n"
    else:
        response += "- No specific precautions found in the dataset for this disease."

    response += "\n**Important Note:** This is an AI-based prediction based on the provided data. It is not a substitute for professional medical advice. Please consult a healthcare professional for diagnosis and treatment."

    return response

# Example Usage (simulating user input)
user_symptoms = ['itching', 'skin_rash', 'nodal_skin_eruptions']
predicted_info = predict_disease_and_precautions(user_symptoms, model, all_symptoms, disease_precautions)
print(predicted_info)
```
